## **Mining Process**

```python
movie_list = []
for i in range(0,774):
    movie_list.append([str(main_df.columns[j]) for j in range(0,1579) if main_df.values[i,j] != 0 ])
```

Code tersebut membuat sebuah list movie_list yang berisi daftar film yang telah diberi rating oleh setiap user pada main_df. Pada setiap iterasi loop, kode tersebut mengecek baris i pada main_df (yaitu user ke-i) dan mencari sel pada setiap kolom j (yaitu judul film ke-j). Jika rating pada sel tersebut tidak sama dengan 0, artinya user i memberikan rating pada film tersebut, maka judul film tersebut ditambahkan ke dalam list movie_list untuk user i.

Penjelasan variabel pada code:

movie_list: list kosong yang akan diisi dengan daftar film yang telah diberi rating oleh setiap user pada main_df.

main_df: dataframe yang digunakan sebagai sumber data.

i: variabel iterasi loop pertama, merepresentasikan indeks baris pada dataframe main_df.

j: variabel iterasi loop kedua, merepresentasikan indeks kolom pada dataframe main_df.

str(main_df.columns[j]): mengambil nama kolom pada indeks ke-j sebagai string.

main_df.values[i,j]: mengambil nilai pada sel baris i dan kolom j dari dataframe main_df.

```python
rules = apriori(transactions = movie_list,min_support=0.05,min_confidence=0.7,min_lift=6,max_length=2)
results = list(rules)
```

Code tersebut melakukan proses mining asosiasi dengan menggunakan algoritma Apriori pada movie_list. Algoritma Apriori digunakan untuk menemukan aturan asosiasi yang terdapat pada kumpulan data, dengan mempelajari kemunculan item-item dalam setiap transaksi. Dalam hal ini, kumpulan data yang digunakan adalah movie_list, yang berisi daftar film yang telah diberi rating oleh setiap user pada main_df.

Penjelasan variabel pada code:

rules: variabel yang menyimpan hasil proses mining asosiasi dengan algoritma Apriori. transactions mengacu pada data input yang digunakan, yaitu movie_list.

min_support adalah nilai support minimum yang digunakan untuk memfilter aturan asosiasi yang terbentuk. Disini **minimal supportnya adalah 5%.**

min_confidence adalah nilai confidence minimum yang digunakan untuk memfilter aturan asosiasi yang terbentuk. Disini **minimal confidence adalah 20%.**

min_lift adalah nilai lift minimum yang digunakan untuk memfilter aturan asosiasi yang terbentuk. Disini **minimal lift untuk memfilter adalah 6.**

max_length adalah panjang maksimum aturan asosiasi yang ingin dicari.

results: variabel yang menyimpan hasil mining asosiasi dalam bentuk list.

Hasil mining asosiasi berisi aturan asosiasi yang terbentuk dan beberapa metrik yang digunakan untuk mengukur kualitas aturan tersebut, seperti support, confidence, dan lift. Setiap aturan asosiasi direpresentasikan sebagai frozenset (set yang tidak dapat diubah) dari item-item yang terlibat dalam aturan tersebut.

```python
def data_extractor(data):
    m1 = np.array([tuple(i[2][0][0] for i in data)]).T
    m1 = pd.DataFrame(m1,columns=['Movie_Watched'])
    m2 = np.array([tuple(i[2][0][1] for i in data)]).T
    m2 = pd.DataFrame(m2,columns=['Also_Watched'])
    support = pd.DataFrame([i[1] for i in data],columns=['Support'])
    confidence = pd.DataFrame([i[2][0][2] for i in data],columns=['Confidence'])
    lift = pd.DataFrame([i[2][0][3] for i in data],columns=['Lift'])
    result = pd.concat([m1,m2,support,confidence,lift],axis=1)
    return result
result_df = data_extractor(results)
```

Code ini adalah sebuah fungsi Python bernama "data_extractor" yang mengambil satu parameter "data". Fungsi ini menggunakan NumPy dan Pandas untuk melakukan ekstraksi data dari variabel "data" yang diinputkan.

Pertama-tama, fungsi ini membuat sebuah array NumPy bernama "m1" yang berisi data dari kolom ke-3 (indeks ke-2) dari "data". Data tersebut diambil dari indeks pertama (indeks ke-0) dari item dalam "data". Kemudian, data tersebut diubah ke dalam bentuk transpose (T) dan dijadikan sebagai DataFrame Pandas dengan nama kolom "Movie_Watched".

Selanjutnya, fungsi ini melakukan hal yang sama untuk membuat array NumPy "m2" yang berisi data dari kolom ke-3 dari "data", tetapi diambil dari indeks kedua (indeks ke-1) dari item dalam "data". Data juga diubah ke dalam bentuk transpose (T) dan dijadikan sebagai DataFrame Pandas dengan nama kolom "Also_Watched".

Selanjutnya, fungsi ini membuat tiga DataFrame Pandas baru: "support" yang berisi data dari kolom ke-1 (indeks ke-0) dari "data", "confidence" yang berisi data dari indeks ke-2 dari "data" dengan indeks ke-0 yang berisi confidence, dan "lift" yang berisi data dari indeks ke-2 dari "data" dengan indeks ke-1 yang berisi lift.

Setelah itu, fungsi ini menggabungkan semua DataFrame yang telah dibuat sebelumnya menjadi satu DataFrame baru dengan menggunakan fungsi "concat" dari Pandas. DataFrame ini dijadikan sebagai output fungsi "data_extractor".

Kemudian, hasil output dari fungsi "data_extractor" disimpan dalam variabel "result_df" dengan memanggil fungsi tersebut dan melewatkan hasil ekstraksi data sebagai argumen.

```python
result_df.sort_values(by='Lift',ascending=False)
```
